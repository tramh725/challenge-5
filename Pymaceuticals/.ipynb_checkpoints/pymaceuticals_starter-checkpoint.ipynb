```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
origin_df = pd.merge(study_results, mouse_metadata, on="Mouse ID", how="left")

# Display the data table for preview
origin_df.head()
```

```python
# Checking the number of mice
mice = origin_df["Mouse ID"].nunique()
mice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = origin_df.loc[origin_df.duplicated(subset=["Mouse ID", "Timepoint"]), "Mouse ID"].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice_df = origin_df.loc[origin_df["Mouse ID"] == "g989", :]
duplicate_mice_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_df = origin_df[origin_df["Mouse ID"].isin(duplicate_mice)==False]
cleaned_df
```

```python
# Checking the number of mice in the clean DataFrame.
cleaned_mice = cleaned_df["Mouse ID"].nunique()
cleaned_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
mean_tumor = cleaned_df.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]
median_tumor = cleaned_df.groupby("Drug Regimen").median()["Tumor Volume (mm3)"]
var_tumor = cleaned_df.groupby("Drug Regimen").var()["Tumor Volume (mm3)"]
std_dev_tumor = cleaned_df.groupby("Drug Regimen").std()["Tumor Volume (mm3)"]
std_err_tumor = cleaned_df.groupby("Drug Regimen").sem()["Tumor Volume (mm3)"]

summary_df = pd.DataFrame({"Mean Tumor Volume": mean_tumor, "Median Tumor Volume": median_tumor, "Tumor Volume Variance": var_tumor,
                  "Tumor Volume Std. Dev.": std_dev_tumor, "Tumor Volume Std. Err.": std_err_tumor})
summary_df
```

```python
agg_summary_df = cleaned_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].agg(['mean', 'median', 'var', 'std', 'sem'])
agg_summary_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
max_treatments = cleaned_df.groupby(["Mouse ID"])["Timepoint"].max()
max_treatments

max_treatments = pd.DataFrame(max_treatments)
max_treatments = max_treatments.reset_index()
max_treatments_df = pd.merge(max_treatments, cleaned_df, on=("Mouse ID", "Timepoint"), how="left")
max_treatments_df

max_capomulin_df = max_treatments_df.loc[max_treatments_df["Drug Regimen"] == "Capomulin",:]
max_ramicane_df = max_treatments_df.loc[max_treatments_df["Drug Regimen"] == "Ramicane", :]
max_infubinol_df = max_treatments_df.loc[max_treatments_df["Drug Regimen"] == "Infubinol", :]
max_ceftamin_df = max_treatments_df.loc[max_treatments_df["Drug Regimen"] == "Ceftamin", :]
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# Create empty list to fill with tumor vol data (for plotting)
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    # Locate the rows which contain mice on each drug and get the tumor volumes
    # add subset 
    # Determine outliers using upper and lower bounds
treatment_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
treatment_vol = []

for treatment in treatment_list:
    location=max_treatments_df.loc[max_treatments_df["Drug Regimen"] == treatment, "Tumor Volume (mm3)"]
    treatment_vol.append(treatment_list)
    
    quartiles = location.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    
    outlier = location.loc[(location<lower_bound)|(location>upper_bound)]
    print(f"{treatment}'s potential outliers: {outlier}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig1, ax1 = plt.subplots()
ax1.set_ylabel('Final Tumor Volume (mm3)')
ax1.boxplot(treatment_vol,0,'ro',labels=treatment_list)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
l509 = cleaned_df.loc[cleaned_df["Mouse ID"]=="l509" , :]
l509

l509_time = l509["Timepoint"]
l509_vol = l509["Tumor Volume (mm3)"]

plt.plot(l509_time, l509_vol)
plt.title("Capomulin Treatment of Mouse l509")
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_df = cleaned_df.loc[cleaned_df["Drug Regimen"]=="Capomulin", :]
capomulin_df = capomulin_df.groupby(["Mouse ID"]).mean()

capo_weight = capomulin_df["Weight (g)"]
capo_vol = capomulin_df["Tumor Volume (mm3)"]

plt.scatter(capo_weight, capo_vol)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
capo_weight = capomulin_df["Weight (g)"]
capo_vol = capomulin_df["Tumor Volume (mm3)"]
plt.scatter(capo_weight, capo_vol)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

pe_slope, pe_int, pe_r, pe_p, pe_std_err = st.linregress(capo_weight,capo_vol)
print(f"The correlation between mouse weight and the average tumor volume is {pe_r}")
regress_values = capo_weight * pe_slope + pe_int
plt.plot(capo_weight, regress_values, color="r")

plt.show()
```
